```python
# beta 8.0.1
import pickle
path = "C:/Users/user/Desktop/MLGame/games/snake/log/t (1).pickle"
file = open(path, "rb")
data = pickle.load(file)
info = data['ml']['scene_info']
game_command = data['ml']['command']
file.close()
for i in range(2, 180):
    path = "C:/Users/user/Desktop/MLGame/games/snake/log/t (" + str(i) + ").pickle"
    file = open(path, "rb")
    data = pickle.load(file)
    info = info + data['ml']['scene_info']
    game_command = game_command + data['ml']['command']
    file.close()

print(type(info[0]))
print(info[0])
print(len(game_command))
```

```python
# 提取特徵
snakehead_x = []
snakehead_y = []
prev_state = []
food_x = []
food_y = []
snaketail_x = []
snaketail_y = []
command = []
up = []
down = []
left = []
right = []
prev_c = 2

for i in range(1, len(info)-2):
    scene_info = info[i]
    snake_head = scene_info["snake_head"]
    food = scene_info["food"]
    snake_body = scene_info["snake_body"]
    length = len(snake_body)
    u = 0
    d = 0
    l = 0
    r = 0
    c = game_command[i]
    if (snake_head[0], snake_head[1] - 10) in snake_body or snake_head[1] - 10 < 0:
        u = 1
    if (snake_head[0], snake_head[1] + 10) in snake_body or snake_head[1] + 10 >= 300:
        d = 1
    if (snake_head[0] - 10, snake_head[1]) in snake_body or snake_head[0] - 10 < 0:
        l = 1
    if (snake_head[0] + 10, snake_head[1]) in snake_body or snake_head[0] + 10 >= 300:
        r = 1
            
    if c == "UP":
        snakehead_x.append(snake_head[0])
        snakehead_y.append(snake_head[1])
        prev_state.append(prev_c)
        food_x.append(food[0])
        food_y.append(food[1])
        snaketail_x.append(snake_body[length-1][0])
        snaketail_y.append(snake_body[length-1][1])
        up.append(u)
        down.append(d)
        left.append(l)
        right.append(r)
        command.append(1)
        prev_c = 1
    elif c == "LEFT":
        snakehead_x.append(snake_head[0])
        snakehead_y.append(snake_head[1])
        prev_state.append(prev_c)
        food_x.append(food[0])
        food_y.append(food[1])
        snaketail_x.append(snake_body[length-1][0])
        snaketail_y.append(snake_body[length-1][1])
        up.append(u)
        down.append(d)
        left.append(l)
        right.append(r)
        command.append(3)
        prev_c = 3
    elif c == "RIGHT":
        snakehead_x.append(snake_head[0])
        snakehead_y.append(snake_head[1])
        prev_state.append(prev_c)
        food_x.append(food[0])
        food_y.append(food[1])
        snaketail_x.append(snake_body[length-1][0])
        snaketail_y.append(snake_body[length-1][1])
        up.append(u)
        down.append(d)
        left.append(l)
        right.append(r)
        command.append(4)
        prev_c = 4
    elif c == "DOWN":
        snakehead_x.append(snake_head[0])
        snakehead_y.append(snake_head[1])
        prev_state.append(prev_c)
        food_x.append(food[0])
        food_y.append(food[1])
        snaketail_x.append(snake_body[length-1][0])
        snaketail_y.append(snake_body[length-1][1])
        up.append(u)
        down.append(d)
        left.append(l)
        right.append(r)
        command.append(2)
        prev_c = 2
print(len(food_x))
print(len(command))
```

```python
import numpy as np

numpy_data = np.array([snakehead_x, snakehead_y, prev_state, food_x, food_y, snaketail_x, snaketail_y, 
                       up, down, left, right])
X = np.transpose(numpy_data) 
y = command
```
