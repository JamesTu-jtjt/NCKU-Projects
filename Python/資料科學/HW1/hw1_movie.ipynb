```python
import pandas as pd
#read file
data = pd.read_csv('IMDB-Movie-Data.csv')
#modify dataframe for further use
#rename runtime and revenue
data.rename(columns = {'Runtime (Minutes)': 'Runtime', 'Revenue (Millions)': 'Revenue'}, inplace = True)
#initialization of lists for further use
actors_d = []
actors = []
directors = []
genres = []
#splitting actors and genres by |
for i in range(len(data)):
    a = data['Actors'].iloc[i]
    g = data['Genre'].iloc[i]
    d = data['Director'].iloc[i]
    tmp_a = a.split('|')
    for i in range(len(tmp_a)):
        tmp_a[i] = tmp_a[i].strip(' ')
        if tmp_a[i].strip() not in actors_d:
            actors_d.append(tmp_a[i].strip())
    actors.append(tmp_a)
    tmp_g = g.split('|')
    for i in range(len(tmp_g)):
        tmp_g[i] = tmp_g[i].strip()
    genres.append(tmp_g)
    if d not in directors:
        directors.append(d)
#add result to dataframe
data['Actors List'] = actors
data['Genres List'] = genres
#show modified dataframe
data.fillna(0, inplace = True)
#data
```

## problem 2

```python
#initialization of list for average revenue
ave_rev = []
#use loop to find and calculate average revenue
for i in actors_d:
    tmp = 0
    rev = 0
    m = data[data['Actors List'].str.contains(i, regex = False)]
    for j in range(len(m)):
        tmp +=1
        rev += m['Revenue'].iloc[j]       
    ave_rev.append(rev/tmp)
#use new dataframe to store results and find top 1
df2 = pd.DataFrame({'Actor': actors_d, 'Average Revenue': ave_rev})
#sort
df2 = df2.sort_values('Average Revenue', ascending = False).set_index('Actor')
#show result
df2.head(2)
```

```python
#initialization
tmp = 0
total_r = 0
#find all movies with E.W. and add to tmp and total_r
m = data[data['Actors List'].str.contains('Emma Watson', regex = False)]
for i in range(len(m)):
    tmp += 1
    total_r += m['Rating'].iloc[i]
#calculate average rating
ave_rating = total_r/tmp
#print result
print("Average Rating of Emma Watson's movies is: " + str(ave_rating))
```

## problem 4

```python
#initialization
d_collabs = []
#find num of collabs and append result
for i in directors:
    c = data[data['Director'] == i]
    collabs = []
    for j in range(len(c)):
        for k in c['Actors List'].iloc[j]:
            if k not in collabs:
                collabs.append(k)
    d_collabs.append(len(collabs))
#store result in dataframe
df4 = pd.DataFrame({'Director': directors, 'Total Collaborations': d_collabs})
df4 = df4.sort_values('Total Collaborations', ascending = False).set_index('Director')
df4.head(4)
```

## problem 5

```python
#initialization of list for total genres
a_genre = []
#use loop to find and calculate average revenue
for i in actors_d:
    tmp = []
    m = data[data['Actors List'].str.contains(i, regex = False)]
    for j in range(len(m)):
        for k in m['Genres List'].iloc[j]:
            if k not in tmp:
                tmp.append(k)
    a_genre.append(len(tmp))
#use new dataframe to store results and find top 1
df5 = pd.DataFrame({'Actor': actors_d, 'Number of Genres': a_genre}).set_index('Actor')
#sort
df5.sort_values('Number of Genres', ascending = False, inplace = True)
#show result
df5.head(6)
```

## problem 6

```python
#initialization of list for gap years
gap = []
#use loop to find and calculate average revenue
for i in actors_d:
    tmp = [2021, 0]
    m = data[data['Actors List'].str.contains(i, regex = False)]
    for j in range(len(m)):
        if m['Year'].iloc[j] < tmp[0]:
            tmp[0] = m['Year'].iloc[j]
        if m['Year'].iloc[j] > tmp[1]:
            tmp[1] = m['Year'].iloc[j]
    gap.append(tmp[1] - tmp[0])
df6 = pd.DataFrame({'Actor': actors_d, 'Gap Years': gap})
#sort
df6 = df6.sort_values('Gap Years', ascending = False).set_index('Actor')
#show result
print("The total number of results is: " + str(len(df6[df6['Gap Years'] == 10])))
df6[df6['Gap Years'] == 10]
```

## problem 7

```python
#write recursive function to accomplish this problem
def find_Collab(df, target, indirect = True):
    #initialization of list to check for updated collaborations
    updated = []
    #find collaborations within target list and append to updated
    for i in target:
        m = df[df['Actors List'].str.contains(i, regex = False)]
        for j in range(len(m)):
            for k in m['Actors List'].iloc[j]:
                if (k not in target) and (k not in updated):
                    updated.append(k)
    #if default find total direct and indirect collaborations through recursive and stop when no collaborations are updated
    if indirect:
        if len(updated)>0:
            #update target and feed back into function
            target.extend(updated)
            return find_Collab(df, target)
        else:
            #print result and return collaborations with target actor as target index 0
            print(target[0], "total direct and indirect collaborations: ", len(target)-1)
            return target
    #else if indirect is false return direct collaborations with target actor as target index 0
    else: 
        target.extend(updated)
        #print result and return collaborations with target actor as target index 0
        print(target[0], "total direct collaborations: ", len(target)-1)
        return target
jd_direct = find_Collab(data, ['Johnny Depp'], indirect = False)
print(jd_direct, '\n')
jd_indirect = find_Collab(data, ['Johnny Depp'])
print(jd_indirect)
```
